```python
import numpy as np
import matplotlib.pyplot as plt

def rc_circuit_visualization(R, C, V_in, t_max, num_points, mode="charging"):
    """
    Visualize the voltage across the capacitor in an RC circuit.

    Parameters:
        R (float): Resistance in ohms (Ω)
        C (float): Capacitance in farads (F)
        V_in (float): Input voltage in volts (V)
        t_max (float): Maximum time for simulation in seconds
        num_points (int): Number of points for the time axis
        mode (str): "charging" or "discharging"
    """
    # Time axis
    t = np.linspace(0, t_max, num_points)
    
    # Calculate voltage across the capacitor
    if mode == "charging":
        V_c = V_in * (1 - np.exp(-t / (R * C)))  # Charging equation
    elif mode == "discharging":
        V_c = V_in * np.exp(-t / (R * C))  # Discharging equation
    else:
        raise ValueError("Invalid mode. Use 'charging' or 'discharging'.")
    
    # Plot the results
    plt.figure(figsize=(8, 5))
    plt.plot(t, V_c, label=f"RC Circuit ({mode.capitalize()})", color="blue")
    plt.title(f"RC Circuit - {mode.capitalize()} Mode", fontsize=14)
    plt.xlabel("Time (s)", fontsize=12)
    plt.ylabel("Voltage Across Capacitor (V)", fontsize=12)
    plt.grid(True)
    plt.legend()
    plt.show()

# Example usage
if __name__ == "__main__":
    R = 1000  # Resistance in ohms (1 kΩ)
    C = 0.001  # Capacitance in farads (1 mF)
    V_in = 5  # Input voltage in volts (5 V)
    t_max = 5  # Maximum time in seconds
    num_points = 1000  # Number of points for the time axis

    # Visualize charging
    rc_circuit_visualization(R, C, V_in, t_max, num_points, mode="charging")

    # Visualize discharging
    rc_circuit_visualization(R, C, V_in, t_max, num_points, mode="discharging")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ======== CHANGE THIS TO YOUR DOWNLOAD LOCATION ========
file_path = r"C:\Users\aswin\Downloads\swing_data (4).csv"

# Load CSV
df = pd.read_csv(file_path)

# Rename columns if needed
df.columns = [c.strip() for c in df.columns]

# Detect correct column names
speed_col = [c for c in df.columns if "Speed" in c][0]
impact_col = [c for c in df.columns if "Impact" in c][0]
angle_col = [c for c in df.columns if "Angle" in c][0]

# Convert to numeric
df[speed_col] = pd.to_numeric(df[speed_col], errors="coerce")
df[impact_col] = pd.to_numeric(df[impact_col], errors="coerce")
df[angle_col] = pd.to_numeric(df[angle_col], errors="coerce")

# Remove invalid rows
df = df.dropna()

# ======== SUMMARY STATS ========
top_speed = df[speed_col].max()
avg_speed = df[speed_col].mean()

top_impact = df[impact_col].max()
avg_impact = df[impact_col].mean()

top_angle = df[angle_col].max()
avg_angle = df[angle_col].mean()

total_swings = len(df)

# ======== PLOTS ========
plt.figure(figsize=(10,5))
plt.plot(df[speed_col], linewidth=2)
plt.title("Swing Speed Trend (km/h)")
plt.xlabel("Swing Number")
plt.ylabel("Speed (km/h)")
plt.grid(True)
plt.show()

plt.figure(figsize=(10,5))
plt.plot(df[angle_col], linewidth=2)
plt.title("Peak Angular Velocity Trend (°)")
plt.xlabel("Swing Number")
plt.ylabel("Angle (°)")
plt.grid(True)
plt.show()

plt.figure(figsize=(10,5))
plt.plot(df[impact_col], linewidth=2)
plt.title("Impact Force Trend (m/s²)")
plt.xlabel("Swing Number")
plt.ylabel("Impact (m/s²)")
plt.grid(True)
plt.show()

# ======== PRINT SUMMARY ========
print("\n========= SWING PERFORMANCE SUMMARY =========")
print(f"Total Swings Recorded     : {total_swings}")
print("---------------------------------------------")
print(f"Top Swing Speed           : {top_speed:.2f} km/h")
print(f"Average Swing Speed       : {avg_speed:.2f} km/h")
print("---------------------------------------------")
print(f"Highest Impact Force      : {top_impact:.2f} m/s²")
print(f"Average Impact Force      : {avg_impact:.2f} m/s²")
print("---------------------------------------------")
print(f"Maximum Swing Angle       : {top_angle:.2f}°")
print(f"Average Swing Angle       : {avg_angle:.2f}°")
print("=============================================\n")
```
